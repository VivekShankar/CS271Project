## VinBigData Dataset

```python
vinbigdata_processed_path = "VinBigData-Chest-X-ray-Abnormalities-Detection/vinbigdata_classifierPP_2021/input/vinbigdata-chest-xray-abnormalities-detection"
vinbigdata_train_df = pd.read_csv(os.path.join(vinbigdata_processed_path, "train.csv"))

vinbigdata_train_df.head()
```

```python
set(vinbigdata_train_df["class_name"])
```

#### Get class_name_to_id dict

```python
class_name_to_id = dict()
for i, row in vinbigdata_train_df.iterrows():
    class_name, class_id = row["class_name"], row["class_id"]
    if class_name in class_name_to_id:
        assert(class_name_to_id[class_name] == class_id)
    else:
        class_name_to_id[class_name] = class_id
print(class_name_to_id)
```
